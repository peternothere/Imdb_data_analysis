```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
imdb_data = pd.read_csv('dataset_imdb/imdb_top_1000.csv')
imdb_data
```

```python
#Description of the dataset
imdb_data.describe(include='all')
```

```python
imdb_data.columns
```

```python
#top 10 movies by imdb rating
top_10 = ['Series_Title','Released_Year','IMDB_Rating']
print("Top 10 movies are:\n",imdb_data[top_10].head(10))
```

```python
#top 10 action movies
action_movies = imdb_data[imdb_data['Genre'].str.contains('Action')]
top_10_action = action_movies.nlargest(10, 'IMDB_Rating')
print('Top action movies are:\n',top_10_action[top_10].set_index('Series_Title'))
```

```python
#top 10 romantic movies
romantic_movies = imdb_data[imdb_data['Genre'].str.contains('Romance')]
top_10_romance = romantic_movies.nlargest(10, 'IMDB_Rating')
print('Top romantic movies are:\n',top_10_romance[top_10].set_index('Series_Title'))
```

```python
#Name all the christopher nolan movies
nolan_movie = []
nolan_movie_full = []
for index, row in imdb_data.iterrows():
    if row['Director'] == 'Christopher Nolan':
        nolan_movie.append(row['Series_Title'])
        nolan_movie_full.append(row)
print("Christopher Nolan movies are:\n",nolan_movie)
```

```python
#christopher nolan movie with the most ratings
df = pd.DataFrame(nolan_movie_full)
sns.barplot(x='IMDB_Rating',y='Series_Title',data=df)
plt.title("Top Nolan Movies")
plt.show()
```
